Just as with scripts by default, python will not reload package modules when code changes.   

The autoreload can also be used with editable packages.    

Without autoreload, reloading code changes requires restarting the ipython kernel.  

Using both a develop editable package install and autoreload, you can make changes to your package and testing it.  
Install package in develop mode. 

    pip install -e ./mypackage
    

Prefixing a command by *!* sends an operating system terminal command instead of a python command

```python
!pip install -e ./mypackage
```

```python
%load_ext autoreload
%autoreload 2

import mypackage
from mypackage import foo, MyClass, myplot
import numpy as np
```

```python
help(mypackage.MyClass)
```

```python
t = np.arange(-1, 1, 0.01)
y = foo(t, amplitude=1, period=0.2)
myplot(t, y)
```

```python
t = np.arange(-1, 1, 0.01)
obj = MyClass(amplitude=1, period=0.2)
y = obj.foo(t)
myplot(t, y)
```
